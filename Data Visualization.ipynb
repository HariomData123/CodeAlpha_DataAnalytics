```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('ipl auction stats.csv')
```

```python
# Display basic info about the data
print("Data Info:")
print(df.head())
print(f"\nShape: {df.shape}")
print(f"Columns: {list(df.columns)}")
```

```python
# Clean the data (assuming IPL data structure)
# Remove ₹ symbol and convert to numeric
if 'FUNDS REMAINING' in df.columns:
    df['FUNDS_NUMERIC'] = df['FUNDS REMAINING'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
# Create visualizations with proper sizing
plt.figure(figsize=(18, 12))  # Increased figure size
```

```python
# 1. Bar Chart - Funds Remaining
plt.subplot(2, 3, 1)
plt.bar(df['TEAM'], df['FUNDS_NUMERIC'])
plt.title('Funds Remaining by Team')
plt.xlabel('Teams')
plt.ylabel('Funds (₹)')
plt.xticks(rotation=45, ha='right', fontsize=8)  # Better rotation and alignment
plt.subplots_adjust(bottom=0.5)  # Add space at bottom
```

```python
# 2. Pie Chart - Total Players Distribution
plt.subplot(2, 3, 2)
# Use shortened team names for pie chart
team_short = [team.split()[-1] if len(team.split()) > 1 else team[:8] for team in df['TEAM']]
plt.pie(df['TOTAL PLAYERS'], labels=team_short, autopct='%1.1f%%', textprops={'fontsize': 8})
plt.title('Total Players Distribution')
```

```python
# 3. Line Plot - Team Index vs Funds
plt.subplot(2, 3, 5)
plt.plot(df.index, df['FUNDS_NUMERIC'], marker='o')
plt.title('Funds by Team Index')
plt.xlabel('Team Index')
plt.ylabel('Funds')
```
